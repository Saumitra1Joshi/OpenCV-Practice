```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(flat_chess)
```

```python
flat_chess_gray = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(flat_chess_gray, cmap='gray')
```

```python
gray = np.float32(flat_chess_gray)
corners = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
```

```python
corners = cv2.dilate(corners, None)
```

```python
plt.imshow(corners)
```

```python
flat_chess[corners > corners.max()] = [255, 0, 0]
```

```python
plt.imshow(flat_chess)
```

```python
real_chess = cv2.imread('../DATA/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(real_chess)
```

```python
real_chess_gray = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(real_chess_gray, cmap='gray')
```

```python
corners = cv2.goodFeaturesToTrack(real_chess_gray, 80, 0.01, 10)
```

```python
corners
```

```python
corners = np.int0(corners)
```

```python
corners
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(real_chess, (x, y), 3, (0, 255, 0), -1)
```

```python
plt.imshow(real_chess)
```
